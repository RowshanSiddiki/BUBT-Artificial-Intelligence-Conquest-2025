```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
%matplotlib inline
import seaborn as sns
import warnings
warnings.filterwarnings('ignore')

pd.set_option('display.max_columns', None)
```

```python
train_df = pd.read_csv('train.csv')
train_df.head()
```

```python
import seaborn as sns
import matplotlib.pyplot as plt

plt.figure(figsize=(10,6))


palette = sns.color_palette("pastel")

sns.barplot(
    x='Sex',
    y='Age',
    hue='Drug',
    data=train_df,
    ci=None,
    palette=palette,
    edgecolor='gray',
    linewidth=0.8
)

plt.title("Average Age by Sex and Drug", fontsize=14, weight='bold')
plt.ylabel("Average Age", fontsize=12)
plt.xlabel("Sex", fontsize=12)
plt.legend(title='Drug')
plt.show()
```

```python
import matplotlib.pyplot as plt

# Features to plot
features = ['Sex', 'Ascites', 'Spiders', 'Edema']
colors = ['#FF9999','#66B3FF','#99FF99','#FFCC99','#FFB6C1','#87CEFA']  # pastel-friendly colors

# Create subplots
fig, axes = plt.subplots(2, 2, figsize=(12,10))
axes = axes.flatten()  # flatten to 1D for easy iteration

for i, feature in enumerate(features):
    counts = train_df[feature].value_counts()
    axes[i].pie(
        counts,
        labels=counts.index,
        autopct='%1.1f%%',
        startangle=90,
        colors=colors[:len(counts)],
        wedgeprops={'width':0.4, 'edgecolor':'white'},
        textprops={'fontsize':12, 'weight':'bold'}  # make percentages bold and larger
    )
    axes[i].set_title(f"{feature} Distribution", fontsize=12, weight='bold')

plt.tight_layout()
plt.show()
```

```python
import matplotlib.pyplot as plt
import seaborn as sns

# Features to plot
features = ['Sex', 'Ascites', 'Spiders', 'Edema']

# Create subplots
fig, axes = plt.subplots(2, 2, figsize=(14,10))
axes = axes.flatten()

for i, feature in enumerate(features):
    counts = train_df[feature].value_counts()

    # Generate dynamic pastel colors based on number of categories
    colors = sns.color_palette("pastel", len(counts))

    axes[i].pie(
        counts,
        labels=counts.index,
        autopct='%1.1f%%',
        startangle=90,
        colors=colors,
        wedgeprops={'width':0.4, 'edgecolor':'white'},
        textprops={'fontsize':12, 'weight':'bold'},
        pctdistance=0.95  # move percentage outward to avoid overlap
    )
    axes[i].set_title(f"{feature} Distribution", fontsize=12, weight='bold')

plt.tight_layout()
plt.show()
```
